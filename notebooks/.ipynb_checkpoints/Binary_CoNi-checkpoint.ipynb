```python
# Grid and time step parameters

# 

a = 3.52e-10
b = a/np.sqrt(2) #m
d = a/np.sqrt(3)
d = d
nd = 50

# segregation parameters
T = 873 #K
E_seg = -0.021 # eV/atom
c0 = 0.57

#
D0 = 5e-6 #m^2/s
Q = 241 #kJ/mol
Q_list = np.random.normal(loc=Q,scale=5,size=nd)

dt = 0.1  # time step size /s
nt = 10000  # number of time steps

E_seg_list = np.zeros(nd)
E_seg_list[0] = E_seg
E_seg_list[1] = 0.0*E_seg
E_seg_list[2] = 0.0*E_seg
```

```python
system = binary_model(D0,
           Q_list, # 1d array, same length as nd
           E_seg_list,# 1d array, same length as nd
           T,
           c0,
           d,
           nd, # 1d array
           dt,
           nt,sublattice_fraction=1)
```

```python
plt.draw()
f = plt.figure(figsize=(8,8))

(ax1) = f.subplots(1,1)
[i.set_linewidth(1.5) for i in ax1.spines.values()]

ax1.set_xlim(0.0,1000)
ax1.set_ylim(54,64)
CB_color_cycle = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00','k']
                  
markers = ['o','<','s','d','x','>','v']

ax1.plot(system.t,system.X_layers_vs_t.T[0]*100,
         color='r',mec='r',mfc='None',ls='-',lw=3,label='Layer-1')
ax1.plot(system.t,system.X_layers_vs_t.T[1]*100,
         color='b',mec='b',mfc='None',ls='-',lw=3,label='Layer-2')



overall_comp = np.array([np.mean(x) for x in system.X_layers_vs_t])*100
ax1.plot(system.t,overall_comp,color='k',mec='k',mfc='None',ls='--',lw=3,label='Averaged')


ax1.axhline(63.5,color='orange',mec='k',mfc='None',ls='-',lw=3,label='Layer-1 (Experiment)')


ax1.tick_params(direction='in',axis='both',length=10,right=True,pad=10)
for labelx1 in (ax1.get_xticklabels()):
    labelx1.set_fontproperties(ticks_font)
    
for labely1 in (ax1.get_yticklabels()):
    labely1.set_fontproperties(ticks_font)
    
ax1.set_ylabel('Co concentration (at.%)',fontproperties=ticks_font)
ax1.legend(fontsize=24,frameon=False,loc='center right')
ax1.set_xlabel('Time (s)',fontproperties=ticks_font)
plt.tight_layout()
```
